## Model Building 

```python
model = smf.ols('Profit ~ RDS + Admin + MRKTS', data = startups).fit()
```

```python
def plot_data(startups,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    startups[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(startups[feature],dist='norm',plot = pylab)


plot_data(startups,'Profit')
plt.title('Profit')
```

## Model Validation

### Two Techniques: 1. Collinearity Check & 2. Residual Analysis

```python
# 1) Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables

Rsq_r = smf.ols("RDS ~ Admin + MRKTS", data = startups).fit().rsquared
VIF_r = 1/(1-Rsq_r)

Rsq_a = smf.ols("Admin ~ RDS + MRKTS", data = startups).fit().rsquared
VIF_a = 1/(1-Rsq_a)

Rsq_m = smf.ols("MRKTS ~ RDS + Admin", data = startups).fit().rsquared
VIF_m = 1/(1-Rsq_m)
```

```python
# Put Values in DataFrame Format

df = {'Variables': ['RDS','Admin','MRKTS'],'VIF':[VIF_r,VIF_a,VIF_m]}
VIF_df = pd.DataFrame(df)
VIF_df
```

```python
# 2) Residual Analysis 
# Test for Normality of Residuals (Q-Q Plot) using residual model (model.resid)

sm.qqplot(model.resid,line = 'q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
# Test for Homoscedasticity or Heteroscedasticity

def standard_values(svals) : return (svals - svals.mean())/svals.std()
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardize Fitted Values')
plt.ylabel('Standardize Residual Values')
plt.show()
```

# Model Deletion Diagnostics 
### Two Techniques : 1. Cook's Distance & 2. Leverage Value 

```python
(c,_) = model.get_influence().cooks_distance
c
```

```python
# Plot the Influencers Using the stem Plot
fig = plt.figure(figsize =(18,6))
plt.stem(np.arange(len(startups)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and Value of Influencer 
np.argmax(c) , np.max(c)
```

```python
# Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k = startups.shape[1]
n = startups.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
startups[startups.index.isin([49])]
```

### Improving Model 

```python
startups = startups.drop(startups.index[[49]],axis = 0).reset_index(drop = True)
startups
```

### Model Deletion Diagnostics and Final Model

```python
while np.max(c)>0.5 :
    model = smf.ols("Profit ~ RDS + Admin + MRKTS", data = startups).fit()
    (c,_) = model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    startups = startups.drop(startups.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
else:
    final_model = smf.ols("Profit ~ RDS + Admin + MRKTS", data = startups).fit()
    final_model.rsquared , final_model.aic
    print("Model Accuracy is Improved to",final_model.rsquared)
```
